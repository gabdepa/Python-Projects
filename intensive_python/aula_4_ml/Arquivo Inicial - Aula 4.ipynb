# Projeto Ciência de Dados - Previsão de Vendas

- Nosso desafio é conseguir prever as vendas que vamos ter em determinado período com base nos gastos em anúncios nas 3 grandes redes que a empresa Hashtag investe: TV, Jornal e Rádio
- TV, Jornal e Rádio estão em milhares de reais
- Vendas estão em milhões

#### Importar a Base de dados

```python
# Importar a bse de dados para o Python
import pandas as pd

tabela = pd.read_csv("advertising.csv")
display(tabela)# 1: 1.000.000(1 Milhão)
# TV = tabela["TV"].sum()
# display(f"Valor da soma do investimento em é TV igual á: R${TV}")

# Visualizar a base e fazer os ajustes na base de dados
# OK!
```

#### Análise Exploratória
- Vamos tentar visualizar como as informações de cada item estão distribuídas
- Vamos ver a correlação entre cada um dos itens

```python
# Análise exploratória -> entender como a sua base de dados está se comportando
# CORRELAÇÃO = "->"!
# TV -> Vendas
# Rádio -> Vendas
# Jornal -> Vendas

# matplotlib e o seaborn

import seaborn as sns
import matplotlib.pyplot as plt

corr_tabela = tabela.corr()

# Cria o gráfico da correlação da tabela com Annotations(número nos gráfico) e mudando as cores(cmap= color map)
sns.heatmap(corr_tabela, annot=True, cmap="Wistia")

# Exibe o gráfico
plt.show()

# Exibe a tabela da correlação
corr_tabela
```

#### Com isso, podemos partir para a preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
# Treinar e Testar a I.A. com a base de dados
# 70-90% Treino & 10-30% Teste
# Y -> Quem eu quero prever(Vendas)
# X -> Resto da base de dados(Quem vou usar pra fazer a previsão)

y = tabela["Vendas"]
x = tabela[["TV", "Radio", "Jornal"]]

from sklearn.model_selection import train_test_split
# Divide a base de dados com x de treino, x de teste, y de treino e y de teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2)
```

#### Temos um problema de regressão - Vamos escolher os modelos que vamos usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)

```python
 # Criar a inteligência artificial e fazer as previsões
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

ia_regressaolinear = LinearRegression()
ia_arvoredecisao = RandomForestRegressor()

ia_regressaolinear.fit(x_treino, y_treino)
ia_arvoredecisao.fit(x_treino, y_treino)
```

#### Teste da AI e Avaliação do Melhor Modelo

- Vamos usar o R² -> diz o % que o nosso modelo consegue explicar o que acontece

```python
# testa para ver qual inteligencia artifical é melhor
from sklearn.metrics import r2_score # Importa teste R² para verificar qual modelo tem maior acertividade

previsao_regressaolinear = ia_regressaolinear.predict(x_teste)
previsao_arvoredecisao = ia_arvoredecisao.predict(x_teste)

if r2_score(y_teste, previsao_arvoredecisao) > r2_score(y_teste, previsao_regressaolinear):
    mayara_ia = previsao_arvoredecisao
    display("A I.A. Mayara será do modelo de Árvore de Decisão")
else:
    mayara_ia = previsao_regressaolinear
    display("A I.A. Mayara será do modelo de Regressão Linear")


# display(f"Previsão I.A. Regressão Linear: {r2_score(y_teste, previsao_regressaolinear)}")
# display(f"Previsão I.A. Árvore de Decisão: {r2_score(y_teste, previsao_arvoredecisao)}")
```

#### Visualização Gráfica das Previsões

```python
tabela_auxiliar = pd.DataFrame()
tabela_auxiliar["y_teste"] = y_teste
tabela_auxiliar["Previsões Árvore de Decisão"] = previsao_arvoredecisao
tabela_auxiliar["Previsões Regressão Linear"] = previsao_regressaolinear

plt.figure(figsize=(15, 6))
sns.lineplot(data=tabela_auxiliar)
plt.show()
```
