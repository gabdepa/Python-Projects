- Trata a planilha

```python
### Passo 9: Abrir Planilha
import pandas as pd
import warnings

# Ignora Warnings
with warnings.catch_warnings(record=True):
    warnings.simplefilter("always")
    resumo = pd.read_excel(name, engine="openpyxl", skiprows=[0])
    resumo = resumo.drop("Id", axis=1)
# Passo 10: Formatar Planilha
    # Pega o índice da linha contendo o primeiro valor "TOTAL"
    vendas_total_row = resumo[resumo["Descrição"]== "TOTAL"].index[0]
    recebimento_total_row = resumo[resumo["Descrição"]== "TOTAL"].index[1]
    comissoes_total_row = resumo[resumo["Descrição"]== "TOTAL"].index[2]

    # Pega o índice da linha contendo o valor VENDAS:
    vendas_row = resumo[resumo["Descrição"]== "VENDAS:"].index[0]
    # Concatena os valores
    resumo["Descrição"][vendas_total_row] = resumo["Descrição"][vendas_row] + resumo["Descrição"][vendas_total_row]

    # Pega o índice da linha contendo o valor RECEBIMENTO NA UNIDADE:
    recebimento_row = resumo[resumo["Descrição"]== "RECEBIMENTO NA UNIDADE:"].index[0]
    # Concatena os valores
    resumo["Descrição"][recebimento_total_row] = resumo["Descrição"][recebimento_row] + resumo["Descrição"][recebimento_total_row]

    # Pega o índice da linha contendo o valor COMISSOES SOBRE:
    comissoes_row = resumo[resumo["Descrição"]== "COMISSOES SOBRE:"].index[0]
    # Concatena os valores
    resumo["Descrição"][comissoes_total_row] = resumo["Descrição"][comissoes_row] + resumo["Descrição"][comissoes_total_row]

    # Elimina todas as linhas com algum valor vazio
    resumo = resumo.dropna(how="any", axis=0)

    
# Passo 11: Cálculo
    # Pega o valor de RECEBIMENTO NA UNIDADE:TOTAL, onde as linhas em que a coluna "Descrição" é igual a "RECEBIMENTO NA UNIDADE:TOTAL"
    recebimento_total = resumo.loc[resumo["Descrição"] == 'RECEBIMENTO NA UNIDADE:TOTAL', "Valor"]
    
    # Pega o valor de COMISSOES SOBRE:TOTAL, onde as linhas em que a coluna "Descrição" é igual a "COMISSOES SOBRE:TOTAL"
    comissoes_total = resumo.loc[resumo["Descrição"] == 'COMISSOES SOBRE:TOTAL', "Valor"]

    # Pega o valor de NOTA ACERTO/MATERIAL, onde as linhas em que a coluna "Descrição" é igual a "NOTA ACERTO/MATERIAL"
    na_total = resumo.loc[resumo["Descrição"] == 'NOTA ACERTO/MATERIAL', "Valor"]
    
    # Pega o valor de VALOR  FATURA(a receber), onde as linhas em que a coluna "Descrição" é igual a "VALOR  FATURA(a receber)"
    valor_fatura = float(resumo.loc[resumo["Descrição"] == 'VALOR  FATURA(a receber) ', "Valor"])

    # Cálculo do valor de fatura a receber    
    sum = float(comissoes_total) - float(recebimento_total) - float(na_total)
    # Arrendonda valor para 2 casas decimais
    sum = round(sum,2)
    
    # Pega o número total de Colunas e acrescenta 1
    col_w = resumo.shape[1]+1
    # Pega o número da linha em que o Campo Descrição possui o valor "VALOR FATURA(a receber)"
    row_w = resumo[resumo["Descrição"]== 'VALOR  FATURA(a receber) '].index[0]
        
# Passo 12: Verificação dos Valores
    # Se o valor declarado for igual ao calculado, exibe OK     
    if sum == valor_fatura:
        resumo.at[row_w, col_w] = 'OK'
        
    else:
        resumo.at[row_w, col_w] = 'Not OK'
    
    # Pega o valor de N.DÉBITO/T.COMBUSTIVEL, onde as linhas em que a coluna "Descrição" é igual a "N.DÉBITO/T.COMBUSTIVEL"
    na_combustivel = float(resumo.loc[resumo["Descrição"] == 'N.DÉBITO/T.COMBUSTIVEL', "Valor"])
    
    # Pega o valor de N.DÉBITO/ALUGUEL, onde as linhas em que a coluna "Descrição" é igual a "N.DÉBITO/ALUGUEL"
    na_aluguel = float(resumo.loc[resumo["Descrição"] == 'N.DÉBITO/ALUGUEL', "Valor"])

    # Pega o valor de VALOR NF DEDUZIDA, onde as linhas em que a coluna "Descrição" é igual a "VALOR NF DEDUZIDA"
    nf_deduzida = float(resumo.loc[resumo["Descrição"] == 'VALOR NF DEDUZIDA', "Valor"])
    
    nota = float(comissoes_total) - float(na_combustivel) - float(na_aluguel)

    # Pega o número da linha em que o Campo Descrição possui o valor "VALOR NF DEDUZIDA"
    row_w = resumo[resumo["Descrição"]== 'VALOR NF DEDUZIDA'].index[0]
# Passo 12: Verificação dos Valores
    # Se o valor declarado for igual ao calculado, exibe "Valores Notas Batem" 
    if nota == nf_deduzida:
        resumo.at[row_w, col_w] = 'Valores NF batem'

    else:
        resumo.at[row_w, col_w] = 'Valores Notas não batem"'

    display(resumo)
            
    resumo.to_excel(f"{path}Fat_{UNIDADE}_{QUINZENA}.xlsx", index=False)
```
